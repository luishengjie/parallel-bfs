```python
import numpy as np
from src.load_graph import get_graph, gen_balanced_tree
import networkx as nx
import matplotlib.pyplot as plt
```

```python
A = get_graph(directed=False)
print(A)
```

```python
def show_graph_with_labels(adjacency_matrix):
    rows, cols = np.where(adjacency_matrix == 1)
    edges = zip(rows.tolist(), cols.tolist())
    gr = nx.Graph()
    gr.add_edges_from(edges)
    pos = nx.spring_layout(gr)

    nx.draw(gr, pos, node_size=300, with_labels=True)
    plt.show()
```

```python
show_graph_with_labels(A)
```

```python
A1 = gen_balanced_tree(height=3, branch=2)
print(A1)
show_graph_with_labels(A1)
```
